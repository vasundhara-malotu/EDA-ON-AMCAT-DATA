# removing outliers

```python
# Function to remove outliers based on IQR
def remove_outliers_iqr(dataframe, columns):
    for col in columns:
        Q1 = dataframe[col].quantile(0.25)  # First quartile (25th percentile)
        Q3 = dataframe[col].quantile(0.75)  # Third quartile (75th percentile)
        IQR = Q3 - Q1  # Interquartile range

        # Define bounds for outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Remove outliers
        dataframe = dataframe[(dataframe[col] >= lower_bound) & (dataframe[col] <= upper_bound)]
        
    return dataframe

# Remove outliers from the DataFrame
df_cleaned = remove_outliers_iqr(df, numerical_cols)

# Display the shape of the original and cleaned DataFrame
print(f"Original DataFrame shape: {df.shape}")
print(f"Cleaned DataFrame shape: {df_cleaned.shape}")

# Optionally, check how many outliers were removed
print(f"Number of rows removed: {df.shape[0] - df_cleaned.shape[0]}")
```

# 2)Categorical vs. Numerical 

To perform a bivariate analysis that examines relationships between categorical vs. categorical and categorical vs. numerical 
variables in your dataset, we will use several visualizations and statistical tests, including:

1. **Box Plots** for categorical vs. numerical analysis.
2. **Count Plots** for categorical vs. categorical analysis.
3. **Statistical Tests** like ANOVA for categorical vs. numerical and Chi-square tests for categorical vs. categorical.

Here’s a comprehensive code snippet that performs the analyses mentioned above, focusing on the factors like CollegeGPA,Domain,Designation, CollegeTier, and Salary. 


```python
import scipy.stats as stats

# 1. Categorical vs Numerical: Box Plot for Salary by College Tier
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='CollegeTier', y='Salary')
plt.title('Salary Distribution by College Tier')
plt.xlabel('College Tier')
plt.ylabel('Salary')
plt.show()

# ANOVA Test for College Tier vs Salary
anova_results = stats.f_oneway(
    df[df['CollegeTier'] == 'Tier 1']['Salary'].dropna(),
    df[df['CollegeTier'] == 'Tier 2']['Salary'].dropna(),
    df[df['CollegeTier'] == 'Tier 3']['Salary'].dropna()
)

print(f"ANOVA results (College Tier vs Salary): F-statistic = {anova_results.statistic:.4f}, P-value = {anova_results.pvalue:.4f}")

if anova_results.pvalue < 0.05:
    print("There are significant differences in Salary among different College Tiers.")
else:
    print("There are no significant differences in Salary among different College Tiers.")

# 2. Categorical vs Numerical: Box Plot for Salary by Designation
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Designation', y='Salary')
plt.title('Salary Distribution by Designation')
plt.xlabel('Designation')
plt.ylabel('Salary')
plt.xticks(rotation=45)
plt.show()

# ANOVA Test for Designation vs Salary
anova_designation_results = stats.f_oneway(
    *[group['Salary'].dropna() for name, group in df.groupby('Designation')]
)

print(f"ANOVA results (Designation vs Salary): F-statistic = {anova_designation_results.statistic:.4f}, P-value = {anova_designation_results.pvalue:.4f}")

if anova_designation_results.pvalue < 0.05:
    print("There are significant differences in Salary among different Designations.")
else:
    print("There are no significant differences in Salary among different Designations.")



# Ensure 'Domain' is a numerical column and 'Salary' is a numerical column

# 1. Scatter Plot for Domain vs Salary
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Domain', y='Salary')
plt.title('Scatter Plot: Domain vs Salary')
plt.xlabel('Domain Skills')
plt.ylabel('Salary')
plt.show()

# Calculate Pearson correlation and p-value
corr_domain, p_domain = stats.pearsonr(df['Domain'].dropna(), df['Salary'].dropna())
print(f"Correlation between Domain Skills and Salary: {corr_domain:.4f}, P-value: {p_domain:.4f}")

if p_domain < 0.05:
    print("There is a significant correlation between Domain Skills and Salary.")
else:
    print("There is no significant correlation between Domain Skills and Salary.")


# Ensure 'Degree' is a categorical column and 'Salary' is a numerical column

# 1. Box Plot for Salary by Degree
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Degree', y='Salary')
plt.title('Salary Distribution by Degree')
plt.xlabel('Degree')
plt.ylabel('Salary')
plt.xticks(rotation=45)
plt.show()

# ANOVA Test for Degree vs Salary
anova_degree_results = stats.f_oneway(
    *[group['Salary'].dropna() for name, group in df.groupby('Degree')]
)

print(f"ANOVA results (Degree vs Salary): F-statistic = {anova_degree_results.statistic:.4f}, P-value = {anova_degree_results.pvalue:.4f}")

if anova_degree_results.pvalue < 0.05:
    print("There are significant differences in Salary among different Degrees.")
else:
    print("There are no significant differences in Salary among different Degrees.")
```

# 3)Categorical vs Categorical

```python
# 3. Categorical vs Categorical: Count Plot for Designation by College Tier
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Designation', hue='CollegeTier')
plt.title('Count of Designations by College Tier')
plt.xlabel('Designation')
plt.ylabel('Count')
plt.legend(title='College Tier')
plt.xticks(rotation=45)
plt.show()

# Prepare data for Chi-square test
contingency_table = pd.crosstab(df['Designation'], df['CollegeTier'])
chi2, p_chi2, _, _ = stats.chi2_contingency(contingency_table)

print(f"Chi-Square Test results: Chi2-statistic = {chi2:.4f}, P-value = {p_chi2:.4f}")

if p_chi2 < 0.05:
    print("There is a significant association between Designation and College Tier.")
else:
    print("There is no significant association between Designation and College Tier.")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# Assuming df is your DataFrame

# 1. Count Plot for Degree vs. Designation
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Degree', hue='Designation', dodge=True)
plt.title('Count of Designations by Degree')
plt.xlabel('Degree')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend().remove()  # Remove the legend
plt.show()

# Chi-Square Test for Degree vs. Designation
contingency_degree_designation = pd.crosstab(df['Degree'], df['Designation'])
chi2_degree_designation, p_degree_designation, _, _ = stats.chi2_contingency(contingency_degree_designation)

print(f"Chi-Square Test results (Degree vs Designation): Chi2 = {chi2_degree_designation:.4f}, P-value = {p_degree_designation:.4f}")

if p_degree_designation < 0.05:
    print("There is a significant relationship between Degree and Designation.")
else:
    print("There is no significant relationship between Degree and Designation.")

# 2. Count Plot for College Tier vs. Designation
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='CollegeTier', hue='Designation', dodge=True)
plt.title('Count of Designations by College Tier')
plt.xlabel('College Tier')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend().remove()  # Remove the legend
plt.show()

# Chi-Square Test for College Tier vs. Designation
contingency_college_designation = pd.crosstab(df['CollegeTier'], df['Designation'])
chi2_college_designation, p_college_designation, _, _ = stats.chi2_contingency(contingency_college_designation)

print(f"Chi-Square Test results (College Tier vs Designation): Chi2 = {chi2_college_designation:.4f}, P-value = {p_college_designation:.4f}")

if p_college_designation < 0.05:
    print("There is a significant relationship between College Tier and Designation.")
else:
    print("There is no significant relationship between College Tier and Designation.")

# 3. Count Plot for Gender vs. Designation
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Gender', hue='Designation', dodge=True)
plt.title('Count of Designations by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend().remove()  # Remove the legend
plt.show()

# Chi-Square Test for Gender vs. Designation
contingency_gender_designation = pd.crosstab(df['Gender'], df['Designation'])
chi2_gender_designation, p_gender_designation, _, _ = stats.chi2_contingency(contingency_gender_designation)

print(f"Chi-Square Test results (Gender vs Designation): Chi2 = {chi2_gender_designation:.4f}, P-value = {p_gender_designation:.4f}")

if p_gender_designation < 0.05:
    print("There is a significant relationship between Gender and Designation.")
else:
    print("There is no significant relationship between Gender and Designation.")
```

# Times of India article dated Jan 18, 2019 states that “After doing your Computer Science Engineering if you take up jobs as a Programming An-alyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh graduate.”

```python
from scipy import stats
relevant_roles = ['programmer Analyst', 'software engineer', 'hardware␣engineer', 'associate engineer']
filtered_df = df[df['Designation'].isin(relevant_roles)]
salary_data = filtered_df['Salary']
claimed_mean_salary = 2.75 * 100000 # Convert lakhs to the actual unit (e.g.,␣2.75 lakhs = 275000)
t_stat, p_value = stats.ttest_1samp(salary_data, claimed_mean_salary)
print(f"Mean Salary of Selected Roles: {salary_data.mean():.2f}")
print(f"Claimed Mean Salary: {claimed_mean_salary:.2f}")
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")
alpha = 0.05 # Set significance level
if p_value < alpha:
    print("Reject the null hypothesis: The average salary is significantly␣different from the claimed mean.")

else:
    print("Fail to reject the null hypothesis: There is no significant␣difference between the average salary and the claimed mean.")
```
